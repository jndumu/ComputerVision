```python
import cv2  # OpenCV for image processing
import numpy as np  # NumPy for numerical operations
import seaborn as sns  # Seaborn for enhanced plotting
import matplotlib.pyplot as plt  # Matplotlib for visualization

# Load the image in BGR format (default format for OpenCV)
image = cv2.imread('./data/beach-blue.jpg')

# Convert the BGR image to HSV format for easier analysis of color properties
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Calculate the histogram for the Hue channel
# [0] specifies the channel index (Hue channel), [180] is the number of bins, and [0, 180] is the range for Hue
hue_hist = cv2.calcHist([image_hsv], [0], None, [180], [0, 180]).flatten()

# Calculate the histogram for the Saturation channel
# [1] specifies the channel index (Saturation channel), [256] is the number of bins, and [0, 256] is the range for Saturation
saturation_hist = cv2.calcHist([image_hsv], [1], None, [256], [0, 256]).flatten()

# Calculate the histogram for the Value channel
# [2] specifies the channel index (Value channel), [256] is the number of bins, and [0, 256] is the range for Value
value_hist = cv2.calcHist([image_hsv], [2], None, [256], [0, 256]).flatten()

# Set Seaborn's visual style for the plots
# "whitegrid" adds a grid background, and "muted" palette gives soft colors
sns.set(style="whitegrid", palette="muted", font_scale=1.2)

# Create a figure for plotting and set its size (16x6 inches)
plt.figure(figsize=(16, 6))

# Plot the Hue histogram
plt.subplot(1, 3, 1)  # Create the first subplot (1 row, 3 columns, first plot)
sns.lineplot(x=np.arange(180), y=hue_hist, color='orange')  # Line plot for the Hue histogram
plt.title('Hue Histogram', fontsize=14, fontweight='bold')  # Title for the plot
plt.xlabel('Hue Value', fontsize=12)  # Label for the x-axis
plt.ylabel('Frequency', fontsize=12)  # Label for the y-axis
plt.grid(True)  # Enable grid for better readability

# Plot the Saturation histogram
plt.subplot(1, 3, 2)  # Create the second subplot (1 row, 3 columns, second plot)
sns.lineplot(x=np.arange(256), y=saturation_hist, color='green')  # Line plot for the Saturation histogram
plt.title('Saturation Histogram', fontsize=14, fontweight='bold')  # Title for the plot
plt.xlabel('Saturation Value', fontsize=12)  # Label for the x-axis
plt.ylabel('Frequency', fontsize=12)  # Label for the y-axis
plt.grid(True)  # Enable grid for better readability

# Plot the Value histogram
plt.subplot(1, 3, 3)  # Create the third subplot (1 row, 3 columns, third plot)
sns.lineplot(x=np.arange(256), y=value_hist, color='blue')  # Line plot for the Value histogram
plt.title('Value (Brightness) Histogram', fontsize=14, fontweight='bold')  # Title for the plot
plt.xlabel('Value (Brightness)', fontsize=12)  # Label for the x-axis
plt.ylabel('Frequency', fontsize=12)  # Label for the y-axis
plt.grid(True)  # Enable grid for better readability

# Adjust the layout to ensure subplots are properly spaced
plt.tight_layout()

# Display the plots
plt.show()
```

# Histogram Analysis of the Image (`beach-blue.jpg`)

## Overview
The three histograms represent the **Hue**, **Saturation**, and **Value (Brightness)** distributions in the HSV color space for the image `beach-blue.jpg`. Below is a detailed analysis of each histogram.

---

## 1. Hue Histogram
- **Description**:
  - The **Hue histogram** represents the frequency of pixel colors (mapped as Hue values in the range `0-180`).
  - Peaks at specific Hue values indicate dominant colors in the image.

- **Observation**:
  - There are significant peaks around **100-120**.
    - This range corresponds to colors like cyan and blue in the HSV color space.
  - Minimal frequencies in other ranges, indicating the image is dominated by these hues.

- **Interpretation**:
  - The image primarily contains blue and cyan colors, likely representing the sky, water, or other similar elements.

---

## 2. Saturation Histogram
- **Description**:
  - The **Saturation histogram** represents the intensity of the colors, ranging from `0` (grayscale/no saturation) to `255` (fully saturated colors).

- **Observation**:
  - A sharp peak near **250** indicates a significant portion of the image contains highly saturated colors.
  - Lower values are almost negligible, meaning there are few or no desaturated (grayscale-like) areas.

- **Interpretation**:
  - The colors in the image are vivid and vibrant, with little to no dullness. This makes sense for outdoor scenes like beaches, where colors like blue (sky and water) are usually vivid.

---

## 3. Value (Brightness) Histogram
- **Description**:
  - The **Value histogram** represents the brightness of the pixels, ranging from `0` (completely dark) to `255` (completely bright).

- **Observation**:
  - There are significant peaks between **200-255**, indicating a large portion of the image is bright.
  - Lower brightness values have smaller frequencies, but they still contribute, suggesting there are some shadows or darker regions.

- **Interpretation**:
  - The image is generally bright, as expected for a sunny outdoor scene. However, the presence of some darker pixels indicates areas of shade, contrast, or other elements like darker objects.

---

## Overall Insights
1. **Dominant Colors**:
   - The image is predominantly blue and cyan, aligning with a beach scene that includes a blue sky or ocean.

2. **Vividness**:
   - The high saturation values suggest the colors are vibrant and saturated, giving the image a lively appearance.

3. **Brightness**:
   - The high brightness values indicate the image is well-lit, likely captured under bright daylight conditions.
