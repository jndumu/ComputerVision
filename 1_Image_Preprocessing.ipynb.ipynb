```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create a new RGB image

image_rgb = Image.new(mode='RGB', size=(100, 100), color=(np.random.randint(0,255), np.random.randint(0,255), np.random.randint(0,255)))

# Step 2: Convert to numpy array for manipulation
image_rgbn = np.array(image_rgb)

# Step 3: Apply random colors to specific regions
image_rgbn[:20, :20] = [np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255)]  # Top-left
image_rgbn[20:40, 20:40] = [np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255)]  # Center
image_rgbn[40:100, 40:100] = [np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255)]  # Bottom-right

# Step 4: Convert back to an Image object
image_rgb = Image.fromarray(image_rgbn)

# Step 5: Plot the original and modified images using Matplotlib
plt.figure(figsize=(12, 4))

# Plot original image
plt.subplot(1, 3, 1)
plt.imshow(Image.new(mode='RGB', size=(100, 100), color=(np.random.randint(0,255), np.random.randint(0,255), np.random.randint(0,255))))
plt.title("Original Image")
plt.axis('off')

# Plot modified image
plt.subplot(1, 3, 2)
plt.imshow(image_rgb)
plt.title("Modified Image")
plt.axis('off')

# Step 6: Convert to RGBA and plot
image_rgba = image_rgb.convert('RGBA')
plt.subplot(1, 3, 3)
plt.imshow(image_rgba)
plt.title("Image with Alpha")
plt.axis('off')

# Display all plots
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

# Step 1: Create a blank BGR image
image_bgr = np.zeros((100, 100, 3), dtype='uint8')

# Step 2: Assign random colors to each channel
image_bgr[:, :, 0] = np.random.randint(0, 256)  # Blue channel
image_bgr[:, :, 1] = np.random.randint(0, 256)  # Green channel
image_bgr[:, :, 2] = np.random.randint(0, 256)  # Red channel

# Step 3: Display the BGR image
cv2.namedWindow('BGR Image', cv2.WINDOW_NORMAL)
cv2.imshow('BGR Image', image_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# Step 4: Convert to RGB and display
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
cv2.namedWindow('RGB Image', cv2.WINDOW_NORMAL)
cv2.imshow('RGB Image', image_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# Step 5: Convert to Grayscale and display
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
cv2.namedWindow('Gray Image', cv2.WINDOW_NORMAL)
cv2.imshow('Gray Image', image_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Step 6: Add an alpha channel
alpha_channel = np.random.randint(0, 256,(100,100))  # Random alpha values
image_bgra = np.dstack((image_bgr, alpha_channel))  # Add alpha channel to BGR image
image_bgra.shape

```

```python
import os
from PIL import Image
import numpy as np
import cv2

# Function to calculate file size
def file_size(file):
    file_size = os.path.getsize(file)  # Get file size in bytes
    file_size_kB = round(file_size / 1024, 4)  # Convert to KB
    return file_size_kB  # Return file size in KB

# Ensure the directory exists
os.makedirs('./Temporary', exist_ok=True)

# Create an image with Pillow
image = Image.new('RGB', (100, 100), (255, 100, 0))  # Orange image

# Save the image with different qualities and print file sizes
image.save("./Temporary/image_SAVE_95.jpg", quality=95)
print(f"File size of image_SAVE_95.jpg: {file_size('./Temporary/image_SAVE_95.jpg')} KB")

image.save('./Temporary/image_SAVE_50.jpg', quality=50)
print(f"File size of image_SAVE_50.jpg: {file_size('./Temporary/image_SAVE_50.jpg')} KB")

image.save('./Temporary/image_SAVE.jpg')  # Default quality
print(f"File size of image_SAVE.jpg: {file_size('./Temporary/image_SAVE.jpg')} KB")

# Create a random RGB image with NumPy for OpenCV
image_rgb = np.zeros((100, 100, 3), dtype='uint8')
image_rgb[:, :, 0] = np.random.randint(0, 256)  # Blue channel
image_rgb[:, :, 1] = np.random.randint(0, 256)  # Green channel
image_rgb[:, :, 2] = np.random.randint(0, 256)  # Red channel

# Save the image using OpenCV and print file sizes
cv2.imwrite("./Temporary/imagecv_SAVE_100.jpg", image_rgb, [cv2.IMWRITE_JPEG_QUALITY, 100])
print(f"File size of imagecv_SAVE_100.jpg: {file_size('./Temporary/imagecv_SAVE_100.jpg')} KB")

cv2.imwrite("./Temporary/imagecv_SAVE_50.jpg", image_rgb, [cv2.IMWRITE_JPEG_QUALITY, 50])
print(f"File size of imagecv_SAVE_50.jpg: {file_size('./Temporary/imagecv_SAVE_50.jpg')} KB")

cv2.imwrite("./Temporary/imagecv_SAVE_0.jpg", image_rgb, [cv2.IMWRITE_JPEG_QUALITY, 0])
print(f"File size of imagecv_SAVE_0.jpg: {file_size('./Temporary/imagecv_SAVE_0.jpg')} KB")

cv2.imwrite("./Temporary/imagecv_SAVE.png", image_rgb)  # Default PNG compression
print(f"File size of imagecv_SAVE.png: {file_size('./Temporary/imagecv_SAVE.png')} KB")

cv2.imwrite("./Temporary/imagecv_SAVE_0.png", image_rgb, [int(cv2.IMWRITE_PNG_COMPRESSION), 0])  # Minimum compression
print(f"File size of imagecv_SAVE_0.png: {file_size('./Temporary/imagecv_SAVE_0.png')} KB")

cv2.imwrite("./Temporary/imagecv_SAVE_9.png", image_rgb, [int(cv2.IMWRITE_PNG_COMPRESSION), 9])  # Maximum compression
print(f"File size of imagecv_SAVE_9.png: {file_size('./Temporary/imagecv_SAVE_9.png')} KB")
```
